```python
#definir hipotesis en base a los datos que ofrece el database
#La selección de la hipotesis se basa en un analisis rapido del data set, en el que encontramos las columnas con menos datos nulos y que nos pueden aportar un tema más interesante.
#1. Taking into account, male victims of shark attacks mortality rate is clearly lower than the one from male victims.
#2. One of the motives is the activity the victims were partaking in at the time of the attacks.
#3. Was the time of the day during the attack a factor?
```

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("../PROJECT-1/attacks.csv",encoding='latin1')
df
```

```python
df = df.loc[:, ['Time', 'Sex ','Fatal (Y/N)','Activity']]
df
```

```python
df.columns
```

```python
#consultamos cuantos NaN hay en cada columna 
df.isna().sum()
```

```python
#elimino las filas del data set vacias
null_counts = df.isna().sum(axis=1)
df = df[null_counts <= 1]
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df[df['Sex '].notnull() & df['Fatal (Y/N)'].notnull()]
```

```python
df = df[df['Time'].notnull() & df['Activity'].notnull()]
```

```python
df.isna().sum()
```

```python
df
```

```python
df
```

```python
df = df[(df['Sex '] == 'M') | (df['Sex '] == 'F')]
```

```python
df = df[(df['Fatal (Y/N)'] == 'Y') | (df['Fatal (Y/N)'] == 'N')]
```

```python
df
```

```python
df.isna().sum()
```

```python
#concatenamos las dos columnas que nos interesan para nuestra hipotesis 
df['sex_fatal'] = df['Sex '].astype(str) + '_' + df['Fatal (Y/N)'].astype(str)
df['activity_fatal'] = df['Activity'].astype(str) + '_' + df['Fatal (Y/N)'].astype(str)
```

```python
 df.to_csv('clean_data.csv', index=False)
```

```python
df = df[~df['Sex '].isin(['Y'])]
```

```python
print(df['sex_fatal'].value_counts(ascending=True))
print(df['activity_fatal'].value_counts(ascending=True))
```
